Probability distributions can be described by any of a number of functions other than CDFs, including moment-generating functions (MGFs) and characteristic functions (CFs).

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's look at characteristic functions a bit. Note that this are complex-valued.

```python
def plot_complex_function(ax, func, xlim=(-1, 1)):
    xpts = np.linspace(*xlim, 250)
    ax.plot(xpts, np.real(func(xpts)), color='blue', label='real')
    ax.plot(xpts, np.imag(func(xpts)), color='red', label='imaginary')
    ax.plot(xpts, np.abs(func(xpts)), color='black', label='norm')
    ax.plot(xpts, -np.abs(func(xpts)), color='black')
    ax.axhline(0, c='k', ls=':')
    ax.axvline(0, c='k', ls=':')

    ax.legend()
```

```python
fig, ax = plt.subplots()
plot_complex_function(ax, lambda x: np.exp(1j * x), ( -3, 3))
ax.set_xlabel('t')
ax.set_ylabel('CF')
ax.set_title('Characteristic function of degenerate distribution')
```

```python
def plot_cf(func, label, xlim=(-6, 6)):
    fig, ax = plt.subplots()
    plot_complex_function(ax, func, xlim)
    ax.set_xlabel('t')
    ax.set_ylabel('CF')
    ax.set_title(f'Characteristic function of {label}')
```

```python
plot_cf(lambda x: np.exp(1j * x), 'degenerate distribution')
```

```python
plot_cf(lambda x: np.exp(1j * x), 'degenerate distribution, k=1')
```

```python
plot_cf(lambda x: 0.5 + 0.5*np.exp(1j * x),
        'bernoulli distribution, p=0.5')
```

```python
plot_cf(lambda x: 0.25 + 0.75*np.exp(1j * x),
        'bernoulli distribution, p=0.75',
       (-12, 12))
```

```python
plot_cf(lambda t: (np.exp(1j * t) - 1)/(1j * t),
        'uniform distribution, a=0, b=1',
        (-25, 25))
```
