## K-Means Repeatability
Investigating how hard it is to find the global minimum with different numbers of clusters, dimentions, and points.

```python
from sklearn.cluster import KMeans
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(edgeitems=12)
```

```python
n_trials = 1000
n_pts = 6
n_dims = 2
n_clusters = 3

pts = stats.norm(0, 1).rvs((n_trials, n_pts, n_dims))
km = KMeans(n_clusters=n_clusters, n_init=100)
inertia = []
labels = np.zeros((n_trials, n_pts))
for i in range(n_trials):
    km.fit(pts[i])
    inertia.append(km.inertia_)
    labels[i] = km.labels_
```

```python
np.unique(labels, axis=0, return_counts=True)[1]
```

This makes no sense. They are all different, every time, even for a trivial case.
